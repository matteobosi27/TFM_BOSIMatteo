```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

df_class['DATE-TIME'] = pd.to_datetime(df_class['DATE-TIME'])

df_20220508 = df_class[df_class['DATE-TIME'].dt.date == pd.to_datetime('2022-05-08').date()]

df_resto = df_class[df_class['DATE-TIME'].dt.date != pd.to_datetime('2022-05-08').date()]

mean_20220508 = df_20220508.groupby(df_20220508['DATE-TIME'].dt.hour)['MEAN'].mean()

mean_resto = df_resto.groupby(df_resto['DATE-TIME'].dt.hour)['MEAN'].mean()

plt.figure(figsize=(10, 6))
plt.plot(mean_20220508.index, mean_20220508.values, label='8 de mayo', color='brown', linewidth=1.5)
plt.plot(mean_resto.index, mean_resto.values, label='Domingos habituales', linestyle='--', color='tomato', linewidth=0.8)

mpl.rc('font', family='Calibri', size=12)

plt.xlabel('Hora del dÃ­a', fontsize=12, labelpad=20)
plt.ylabel('Velocidad media', fontsize=12, labelpad=20)
plt.title('Perfil temporal de la velocidad media Ronda del Sur', fontsize=12, pad=20)
plt.xticks(rotation=90, fontsize=10) 
plt.yticks(fontsize=10)
plt.legend(fontsize=12) 

plt.savefig('perfil_tempo_embajadores.jpg', format='jpg', dpi=300, bbox_inches='tight')

plt.show()
```
