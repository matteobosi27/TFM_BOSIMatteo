```python
import pandas as pd
import matplotlib.pyplot as plt

dataframes_and_dates = [
    (df3, ['2022-05-15', '2022-04-24']),
    (df2, ['2022-05-15', '2022-04-24']),
    (df2, '2022-05-08'),
    (df3, '2022-05-08')
]

plt.figure(figsize=(10, 6))

for df, dates in dataframes_and_dates:
    if isinstance(dates, list):
        combined_hourly_means = None
        
        for date in dates:
            df_filtered = df[(df['Fecha'] == date) & (df['type'] == 'Restauration')]
            hourly_means = df_filtered.groupby([df_filtered['Formatted_Source'].dt.hour])['current_popularity'].mean()
            
            if combined_hourly_means is None:
                combined_hourly_means = hourly_means
            else:
                combined_hourly_means += hourly_means
        
        combined_hourly_means /= len(dates)  
        
        label = f'Semanas habituales en un radio de {"1" if df is df3 else "5"}km'
        
        linestyle = 'dashed' if df is df3 else 'dashdot'
        linewidth = 2 if df is df3 else 0.75  
        
    else:
        df_filtered = df[(df['Fecha'] == dates) & (df['type'] == 'Restauration')]
        combined_hourly_means = df_filtered.groupby([df_filtered['Formatted_Source'].dt.hour])['current_popularity'].mean()
        
        if df_filtered['type'].iloc[0] == 'Restauration':
            label = f'8 de mayo en un radio de {"1" if df is df3 else "5"}km'
            linestyle = 'solid'
            linewidth = 3 if df is df3 else 0.75  
        else:
            continue

    plt.plot(combined_hourly_means.index, combined_hourly_means.values, label=label, linestyle=linestyle, linewidth=linewidth, color='#069AF3') 

plt.xlabel('Hora')
plt.ylabel('Media de concurrencia (con base 100)')
plt.title('Evolución horaria de la afluencia de los puntos de interés - Restauración', pad=20, fontname='Calibri')

legend_font = {'family': 'Calibri'}
plt.legend(prop=legend_font)
plt.xticks(range(24), labels=[f"{hour:02}:00" for hour in range(24)], rotation='vertical', fontname='Calibri')
plt.yticks(fontname='Calibri')

plt.savefig('evolucion_horaria_popularidad_restauracion.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

date_range = pd.date_range(start='2022-04-26', end='2022-05-08')
df_filtered_period = df2[df2['Fecha'].isin(date_range) & (df2['node_type'] == 'tourist_attraction')]

df_filtered_rest = df2[~df2['Fecha'].isin(date_range) & (df2['node_type'] == 'tourist_attraction')]

plt.figure(figsize=(10, 6))

hourly_means_period = df_filtered_period.groupby([df_filtered_period['Formatted_Source'].dt.hour])['current_popularity'].mean()

hourly_means_rest = df_filtered_rest.groupby([df_filtered_rest['Formatted_Source'].dt.hour])['current_popularity'].mean()

label_period = 'Entre 26 de abril y 8 de mayo en un radio de 5 km'
label_rest = 'Resto de fechas en un radio de 5 km'

plt.plot(hourly_means_period.index, hourly_means_period.values, label=label_period, linestyle='solid', linewidth=1.5, color='#FAC205')
plt.plot(hourly_means_rest.index, hourly_means_rest.values, label=label_rest, linestyle='dashed', linewidth=2, color='#FAC205')

plt.xlabel('Hora')
plt.ylabel('Media de concurrencia (con base 100)')
plt.title('Comparación de la evolución horaria de la afluencia de puntos de interés - Atracciones turísticas', pad=20, fontname='Calibri')

legend_font = {'family': 'Calibri'}
plt.legend(prop=legend_font)
plt.xticks(range(24), labels=[f"{hour:02}:00" for hour in range(24)], rotation='vertical', fontname='Calibri')
plt.yticks(fontname='Calibri')

plt.savefig('comparacion_evolucion_horaria_popularidad_tourist_normal.png', dpi=300, bbox_inches='tight')

plt.show()
```
