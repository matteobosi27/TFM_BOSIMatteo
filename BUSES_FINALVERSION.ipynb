```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
from matplotlib.font_manager import FontProperties

df = pd.DataFrame()

df_bus = pd.read_csv("XXX.csv")
```

```python
df_bus['Hora'] = pd.to_datetime(df_bus['Hora']).dt.hour
```

```python
def extract_hour(idfranja):
    try:
        return int(idfranja[1:])
    except ValueError:
        return None

df_bus['Idfranja'] = df_bus['Idfranja'].astype(str)
df_bus['Hour'] = df_bus['Idfranja'].apply(extract_hour)
df_filtered = df_bus.dropna(subset=['Hour'])

grouped = df_filtered.groupby(['Línea', 'Hour'])
total_viajeros = grouped['Viajeros'].mean()

fig, ax = plt.subplots(figsize=(12, 6))

line_colors = {
    '123': '#0070FF',
    '23': '#D90347',
    '78': '#38A800',
    'T32': '#A900E6',
    '180': '#E6E600',
    'N12': '#A87000',
}

for linea in total_viajeros.index.get_level_values('Línea').unique():
    df_linea = total_viajeros[linea]
    color = line_colors.get(linea, 'gray')
    ax.plot(df_linea.index, df_linea.values, label=linea, color=color)

ax.set_xlabel('Hora', labelpad=10, fontname='Calibri')
ax.set_ylabel('Número de Viajeros', labelpad=10, fontname='Calibri')
ax.set_title('Número de viajeros por hora entre el 15 de abril y el 15 de mayo del 2022 - Todas las líneas', pad=20, fontname='Calibri')

legend_font = fm.FontProperties(family='Calibri')  
ax.legend(prop=legend_font)

hours = df_filtered['Hour'].unique()
ax.set_xticks(hours)

hour_labels = [f"{hour:02}:00" for hour in hours]
ax.set_xticklabels(hour_labels, rotation='vertical', fontname='Calibri')

plt.yticks(fontname='Calibri')

plt.savefig('numeroviajerospordia_todas_bus', dpi=300, bbox_inches='tight')

plt.show()
```

```python
df_bus['Fecha'] = pd.to_datetime(df_bus['Fecha'])

start_date = pd.Timestamp('2022-04-15')
end_date = pd.Timestamp('2022-05-15')
df_filtered = df_bus[(df_bus['Fecha'] >= start_date) & (df_bus['Fecha'] <= end_date)]

avg_viajeros = df_filtered.groupby(['Línea', 'Fecha'])['Viajeros'].mean().unstack()

normalized_values = avg_viajeros.apply(lambda x: (x - x.min()) / (x.max() - x.min()), axis=1)

plt.figure(figsize=(12, 6))

sns.heatmap(normalized_values, cmap='YlGnBu', cbar=False)

for i, line_name in enumerate(normalized_values.index):
    for j, value in enumerate(normalized_values.loc[line_name]):
        plt.text(j + 0.5, i + 0.5, f'{value:.2f}', ha='center', va='center', rotation='vertical', fontname='Calibri')

plt.xlabel('Día', labelpad=10, fontname='Calibri')
plt.ylabel('Línea', labelpad=10, fontname='Calibri')
plt.title('Concurrencia diaria de las líneas, con base 1 el día de mayor concurrencia', pad=20, fontname='Calibri')

date_objects = df_filtered['Fecha'].dt.date.unique()  

plt.xticks(range(len(date_objects)), [date.strftime('%m-%d') for date in date_objects], fontname='Calibri')  
plt.yticks(fontname='Calibri')

plt.savefig('concurrenciadiaria_base1_bus', dpi=300, bbox_inches='tight')

plt.show()
```

```python
df_bus.sort_values(['Línea', 'Fecha', 'Hora'], inplace=True)

df_max_viajeros = df_bus.groupby(['Línea', 'Fecha'])['Viajeros'].agg(['max', 'idxmax']).reset_index()

df_max_viajeros['Hora_Max_Viajeros'] = df_max_viajeros['idxmax'].apply(lambda x: df_bus.loc[x, 'Hora'])

pivot_table = df_max_viajeros.pivot(index='Línea', columns='Fecha', values='Hora_Max_Viajeros')

plt.figure(figsize=(10, 6))

sns.heatmap(pivot_table, cmap='viridis', annot=False, fmt=".2f", cbar=True)

x_coords = range(len(pivot_table.columns))
y_coords = range(len(pivot_table.index))

for i, y in enumerate(y_coords):
    for j, x in enumerate(x_coords):
        plt.text(x + 0.5, y + 0.5, pivot_table.iloc[y, x], ha='center', va='center', rotation='vertical', color='white', fontname='Calibri')

plt.xlabel('Día', labelpad=10, fontname='Calibri')
plt.ylabel('Línea', labelpad=10, fontname='Calibri')
plt.title('Hora del máximo número de viajeros por línea y día', pad=20, fontname='Calibri')

date_objects = df_filtered['Fecha'].dt.date.unique()  

plt.xticks(range(len(date_objects)), [date.strftime('%m-%d') for date in date_objects], fontname='Calibri')  
plt.yticks(fontname='Calibri')

plt.savefig('horamaximoviajeros_bus', dpi=300, bbox_inches='tight')

plt.show()
```
